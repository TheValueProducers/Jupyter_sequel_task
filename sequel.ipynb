```python
import pandas as pd
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
excel_file = '../mySQL sample database.xlsx'
orders = pd.read_excel(excel_file, sheet_name='classicmodels orders')
orders_details = pd.read_excel(excel_file, sheet_name='classicmodels orderdetails' )
customers = pd.read_excel(excel_file, sheet_name='classicmodels customers' )
product_lines = pd.read_excel(excel_file, sheet_name='classicmodels productlines' )
products = pd.read_excel(excel_file, sheet_name='classicmodels products' )
```

```python
order_query = """
    SELECT orderNumber, SUM(quantityOrdered*priceEach) AS total_order_value 
    FROM orders_details 
    GROUP BY orderNumber;
"""
pysqldf(order_query) #Tổng tiền theo từng đơn hàng
```

```python
customer_query = """
    SELECT customers.customerNumber, 
        customers.customerName, 
        SUM(orders_details.quantityOrdered * orders_details.priceEach) as total_orders_value
    FROM customers
    JOIN orders
    ON customers.customerNumber = orders.customerNumber
    JOIN orders_details
    ON orders.orderNumber = orders_details.orderNumber
    GROUP BY customers.customerNumber, customers.customerName;
"""
pysqldf(customer_query) #Tổng tiền theo từng khách hàng
```

```python
product_lines_query = """
    SELECT pl.productLine, SUM(od.quantityOrdered * od.priceEach) as total_order_value
    FROM product_lines as pl
    JOIN products as p
    ON p.productLine = pl.productLine
    JOIN orders_details as od
    ON od.productCode = p.productCode
    GROUP BY 1;
    
"""
pysqldf(product_lines_query) #Tổng tiền theo từng dòng sản phẩm
```

```python
product_query = """
    SELECT p.productName, SUM(od.quantityOrdered * od.priceEach) AS total_order_value
    FROM products AS p
    JOIN orders_details AS od
    ON od.productCode = p.productCode
    GROUP BY 1;
"""
pysqldf(product_query)
```

```python
country_query = """
    SELECT c.country, SUM(od.quantityOrdered * od.priceEach) AS total_order_value
    FROM customers AS c
    JOIN orders AS o
    ON o.customerNumber = c.customerNumber
    JOIN orders_details AS od
    ON o.orderNumber = od.orderNumber
    GROUP BY 1;
    
"""

pysqldf(country_query) # Tổng tiền theo từng quốc gia
```
